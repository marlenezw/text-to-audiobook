```python
import PyPDF2, pikepdf,pyttsx3

#open, decrypt and read the pdf 
pdf = pikepdf.open("short-story.pdf") 
pdf.save('decrypted-short-story.pdf')
pdf_Reader = PyPDF2.PdfFileReader('decrypted-short-story.pdf') 
```

```python
#initiate the voice
speaker = pyttsx3.init() 

#iterate through the pages you want it to read out to you
for num in range(4,6):
    page = pdf_Reader.getPage(num)
    text = page.extractText()
    speaker.say(text)
    speaker.runAndWait()
```

```python
#if you'd like to save the audio to an mp3 file use
speaker.save_to_file(text, 'audiobook.mp3')
speaker.runAndWait()
```
